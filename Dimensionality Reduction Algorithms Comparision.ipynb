## <a id="pca">PCA</a>

 #### <a id="pca15">PCA with 15 components</a>

```python
pca = PCA(n_components=15,random_state=42) #Apply PCA with two compontents
principalComponents = pca.fit_transform(original_features)
print(principalComponents.shape)
print("Total Percentage of Variance in the data",np.round(sum(pca.explained_variance_ratio_),2))

#The estimated noise covariance following the Probabilistic PCA model from Tipping and Bishop 1999. 
#See “Pattern Recognition and Machine Learning” by C. Bishop, 12.2.1 p. 574 
#or http://www.miketipping.com/papers/met-mppca.pdf. It is required to compute the estimated data covariance and score samples.
print("Noise Variance",pca.noise_variance_)
```

```python
#Plotting the Cumulative Summation of the Explained Variance
plt.figure(figsize=(20,5))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components',fontsize=18)
plt.ylabel('Variance (%)',fontsize=18) #for each component
plt.xticks(np.arange(0, len(pca.explained_variance_ratio_), 1))
plt.title('Al Jazeera Explained Variance as the number of components increase',fontsize=20)
plt.show()
```

```python
#Convert the array to pandas dataframe
pcadf = pd.DataFrame(data=principalComponents, columns = range(principalComponents.shape[1]))
pcadf.insert(loc=0, column='Unnamed: 0', value=df['Unnamed: 0'])
pcadf.to_csv('PCA_15_Components.csv',index=False)
pcadf.head()
```

```python
# Correlation Matrix Heatmap
f, ax = plt.subplots(figsize=(10, 6))
corr = pcadf.iloc[:,1:].corr()
hm = sns.heatmap(round(corr,2), annot=True, ax=ax, cmap="coolwarm",fmt='.2f',linewidths=.05)
f.subplots_adjust(top=0.93)
t= f.suptitle('correlation Between Variables after PCA with 15 components', fontsize=14)
```

```python
print("Mean correlation:",np.round(pcadf.corr().sum().sum()/(pcadf.shape[1]-1),2))
```

## <a id="nmf">Non-Negative Matrix Factorization (NMF)<a/>

NMF can be used for example for dimensionality reduction, topic extraction or source separation.

#### <a id="nmf15">NMF with 15 Components<a/>

```python
model = NMF(n_components=15, random_state=0)
W = model.fit_transform(original_features)
print(W.shape)
# Frobenius norm of the matrix difference, or beta-divergence, between the training data X and the reconstructed data WH from the fitted model.
print("Matrix Reconstruction Error (W*H)",model.reconstruction_err_)
```

```python
nmfdf = pd.DataFrame(data=W, columns = range(W.shape[1]))
nmfdf.insert(loc=0, column='Unnamed: 0', value=df['Unnamed: 0'])
nmfdf.to_csv('NMF_15_Components.csv',index=False)
nmfdf.head()
```

## <a id="lda">LDA</a> 

LDA is a supervised algorithm. It tries to maximize the separability between classes (unlike PCA, which maximizes variance). Since it is a supervised algorithm, it can only be used with labeled data. 

Advantages:

    * Since it is supervised algorithm, mostly it improves the predictive performance of the extracted features.

Disadvantages:

    * Features are not easily Interpretable
    * Requires labeled data
